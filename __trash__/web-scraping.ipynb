```python
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd
```

```python
url = 'https://ec.europa.eu/eurostat/statistics-explained/index.php/Glossary:Country_codes'
html = requests.get(url).content
html[:5000]
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.find('table')
table
```

```python
items = [x.text for x in table.find_all('td')]
clean_items = [i.strip('\n') for i in items]
clean_items
```

```python
countrys = []
country_codes = []

for i in clean_items:
    if i.startswith('('):
        country_codes.append(i[1:-1])
    else:
        countrys.append(i)
```

```python
country_dict = dict(zip(countrys, country_codes))
```

```python
country_dict['Great Britain'] = 'GB'
```

```python
country_dict
```

```python
country_df = pd.DataFrame.from_dict(country_dict, orient='index')
```

```python
country_df = country_df.reset_index()
```

```python
country_df[0][1] = 'GR'
```

```python
country_df = country_df.rename(columns={'index':'country', 0:'country_codes'})
```

```python
country_df
```

```python
df_in_process = pd.read_csv('data/processed_df.csv')
```

```python
df_in_process
```

```python
merging_countries = country_df.merge(df_in_process, left_on='country_codes', right_on='country_code')
```
