```python
import requests
import pandas as pd
import pymysql
import sqlalchemy as db
from sqlalchemy import create_engine
```

```python
response = requests.get('http://api.dataatwork.org/v1/jobs?limit=500')
json_data = response.json()
api_jobs = pd.DataFrame(json_data)
```

```python
sqlitedb_path = 'data/raw_data_project_m1.db'

conn_str = f'sqlite:///{sqlitedb_path}'

engine = create_engine(conn_str)
connection = engine.connect()
metadata = db.MetaData()
```

```python
career_info_table = pd.read_sql_query("SELECT * FROM career_info", engine)
```

```python
job_code_list = career_info_table['normalized_job_code'].unique().tolist()
```

```python
url_list = []
for i in job_code_list:
    url_list.append(f'http://api.dataatwork.org/v1/jobs/{i}')
    if i == 'None':
        pass
```

```python
url_list
```

```python
lst = []
for url in url_list:
    response = requests.get(url)
    json_data = response.json()
    lst.append(json_data)
```

```python
df = pd.DataFrame(lst)
df
```

```python
merged_db = career_info_table.merge(df, left_on='normalized_job_code', right_on='uuid')
```

```python
merged_db
```

```python
column = ['uuid_x', 'title']
```

```python
title_column = merged_db[column]
```

```python
title_column
```

```python
country_info_table = pd.read_sql_query("SELECT * FROM country_info", engine)
country_info_table
```

```python
column2 = ['uuid', 'country_code']
```

```python
country_column = country_info_table[column2]
```

```python
country_column
```

```python
merging_country_title = title_column.merge(country_column, left_on='uuid_x', right_on='uuid')
```

```python
merging_country_title
```

```python
personal_info_table = pd.read_sql_query("SELECT * FROM personal_info", engine)
personal_info_table.head(10)
```

```python
age_column = [i.strip(' years old') for i in personal_info_table['age']]
age_column

personal_info_table['age'] = age_column

personal_info_table
```

```python
def calculating_age(year):
    if len(year) == 4:
        return 2016 - round(int(year))
    else:
        return year

personal_info_table['age'] = personal_info_table['age'].apply(calculating_age)
```

```python
personal_info_table['age'] = personal_info_table['age'].apply(lambda x : int(x) + 4)
```
