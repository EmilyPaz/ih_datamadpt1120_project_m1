```python
import pymysql
import sqlalchemy as db
from sqlalchemy import create_engine
import pandas as pd
```

```python
sqlitedb_path = 'data/raw_data_project_m1.db'

conn_str = f'sqlite:///{sqlitedb_path}'

engine = create_engine(conn_str)
connection = engine.connect()
metadata = db.MetaData()
```

```python
all_tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", engine)
all_tables
```

```python
personal_info_table = pd.read_sql_query("SELECT * FROM personal_info", engine)
personal_info_table.head(10)
```

```python
# i'm going to use age column so here i'm making values simple
age_column = [i.strip(' years old') for i in personal_info_table['age']]
age_column

personal_info_table['age'] = age_column

personal_info_table
```

```python
#some of the values are the birth year so i'm going to calculate the age these people had in 2016 according to the rest of the ages in age column
def calculating_age(year):
    if len(year) == 4:
        return 2016 - round(int(year))
    else:
        return year

personal_info_table['age'] = personal_info_table['age'].apply(calculating_age)
```

```python
personal_info_table
```

```python
#now that we have all values as ages i'm calculating the age these people have now (only 4 years more because 2021 just started)
personal_info_table['age'] = personal_info_table['age'].apply(lambda x : int(x) + 4)
```
